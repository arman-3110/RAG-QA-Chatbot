```python
df_train.fillna('missing', inplace=True)
```

```python
# Making a function to convert eacgh row into a descriptive text
def convert_row_to_text(row):
    return (f"Applicant {row['Loan_ID']}: {row['Gender']}, "
            f"{'Married' if row['Married'] == 'Yes' else 'Unmarried'}, "
            f"{row['Dependents']} dependents, "
            f"{row['Education']}, "
            f"{'Self-Employed' if row['Self_Employed'] == 'Yes' else 'Not Self-Employed'}, "
            f"Applicant Income: {row['ApplicantIncome']}, "
            f"Coapplicant Income: {row['CoapplicantIncome']}, "
            f"Loan Amount: {row['LoanAmount']}, "
            f"Loan Term: {row['Loan_Amount_Term']}, "
            f"Credit History: {row['Credit_History']}, "
            f"Property Area: {row['Property_Area']}, "
            f"Loan Status: {'Approved' if row['Loan_Status'] == 'Y' else 'Not Approved'}.")
```

```python
# Now applying the above function to entire dataset
knowledge_base = df_train.apply(convert_row_to_text, axis=1).to_list()
knowledge_base
```

```python
# Using the Sentence Transfer Model 
from sentence_transformers import SentenceTransformer
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Using the embedder to embend the entire knowledge base
embeddings = embedder.encode(knowledge_base,convert_to_numpy=True, show_progress_bar=True)
```

```python
embeddings
```

```python
import faiss

# Dimension of embeddings
embedding_dim = embeddings.shape[1]

# Build FAISS index (Flat = exact search)
index = faiss.IndexFlatL2(embedding_dim)

# Add embeddings to FAISS index
index.add(embeddings)
```

```python
# Example user question
user_question = "How many applicants got their loan approved?"

# Generate embedding for the question
question_embedding = embedder.encode([user_question])
```

```python
# Number of top matches you want to retrieve
k = 3

# Perform similarity search in FAISS
distances, indices = index.search(question_embedding, k)

# Retrieve the top matching knowledge base texts
retrieved_texts = [knowledge_base[i] for i in indices[0]]
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load FLAN-T5 Small model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")

# Create pipeline for text2text generation
qa_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)
```

```python
# Combine the retrieved texts into one context
context = " ".join(retrieved_texts)

# Prepare the prompt
prompt = f"Question: {user_question} Context: {context}"
```

```python
response = qa_pipeline(prompt, max_length=200, do_sample=False)
answer = response[0]['generated_text']

print("Answer:", answer)
```
